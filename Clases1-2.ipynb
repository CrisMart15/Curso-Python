#**Lista de listas (lista compleja)**

```python
lista_compleja = ['hola', 1, 2, 3, 4, 5, [34, 5, 6], {"1":1,"2":2}, 8.0, 9.0, True, 7+6j]
```

```python
for index in range(0, len(lista_compleja)-1):
  if type(lista_compleja[index]) == list:
    output_list = lista_compleja[index]

print(output_list)
```

 **Funciones**

```python
def get_my_list (lista_compleja):
  for index in range (0, len(lista_compleja)-1):
    if type(lista_compleja[index]) == list:
      output_list=lista_compleja[index]
      return output_list
```

```python
lista_obtenida = get_my_list(lista_compleja)
lista_obtenida
```

```python
#Crear una lista vacía y llenarla automaticamente
lista_fill = []
for i in range(0,20):
  lista_fill.append(i)
lista_fill

```

```python
#Crear lista con while
index = 0
while index < len(lista_fill):
  print (lista_fill[index])
  #index = index + 1
  #Se incrementa un valor en index
  index +=1
```

```python
#Creación de intervalos
intervalo = 4
contador = 0
split_list = []

while contador < len(lista_fill):
  #No existe do-while
  split_list.append(lista_fill[contador:contador+intervalo])
  contador += intervalo

print(split_list)
```
