**Definiciones**

*   *Constante:* Cuando no depende de datos de entrada (n)
*   *Logaritmico:* Reduce los capos de entrada en cada step (Binary Search)
*   *Lineal:* El tiempo incrementa linealmente con los datos de entrada
*   *Quasilineal:* Cuando cada operación en los datos de entrada tiene un tiempo de complejidad logaritmica (Merge sort, Heap Sort)
*   *Cuadrático:* Cuando en cada operación que se realiza se desempeña una operación lineal.
*   *Exponencial:* Cuyando el tiempo de complejidad aumenta al doble en cada adición de datos de entrada (Brute-Force algorithms)

```python
def lista_anidada(seq, num):

    avg = len(seq) / float(num)

    out = []

    last = 0.0



    while last < len(seq):

        out.append(sq[int(last):int(last + avg)])

        last += avg



    return out
```

#**Problema** 

Se tiene una lista de datos desordenada, requerimos encontrar el valor mínimo de toda nuestra lista de datos. Implemente una solución optima para asegurar que nuestros recursos satisfagan nuestra necesidad.

```python
lista = [5,4,2,1,0]

def findMin(alist):
  minimo = alist[0]
  for i in alist: #Nos manda directamente a los valores sin necesidad de poner un índice
    valor_minimo = True
    for j in alist:
      if i > j:
        valor_minimo = False
    if valor_minimo:
      minimo = i
  return minimo
```

```python
print (findMin(lista))
```

```python
def findMax(blist):
  minimo = blist[0]
  for i in blist: #Nos manda directamente a los valores sin necesidad de poner un índice
    valor_maximo = True
    for j in blist:
      if i < j:
        valor_maximo = False
    if valor_maximo:
      maximo = i
  return maximo
```

```python
print (findMax(lista))
```
